# Factor Forge Quickstart

This notebook runs a single-factor backtest on synthetic data.

```python
from factor_forge import BacktestEngine, DataLoader
from factor_forge.data.loader import generate_synthetic_prices
from factor_forge.factors.registry import FactorRegistry

prices = generate_synthetic_prices(["A", "B", "C", "D", "E"], "2015-01-01", "2024-01-01")

registry = FactorRegistry()
registry.load_builtin()
factor = registry.get("momentum_12_1")

engine = BacktestEngine(factor=factor, prices=prices, transaction_cost_bps=10.0)
result = engine.run()
print(result.metrics)
```

```python
from factor_forge.viz.cum_returns import plot_cumulative_returns

fig = plot_cumulative_returns({factor.name: result.returns})
fig.show()
```
